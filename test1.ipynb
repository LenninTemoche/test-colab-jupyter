```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el DataFrame
df = pd.read_csv('panini.csv')

# Convertir la columna 'fecha' a tipo datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# --- Ejemplo de Gráfico 1: Ventas por Producto (Top 10) ---
plt.figure(figsize=(12, 6))
sns.countplot(y='producto', data=df, order=df['producto'].value_counts().index[:10], palette='viridis')
plt.title('Top 10 Productos Más Vendidos')
plt.xlabel('Número de Ventas')
plt.ylabel('Producto')
plt.show()

# --- Ejemplo de Gráfico 2: Ventas por Método de Pago ---
plt.figure(figsize=(10, 6))
sns.countplot(x='metodo_pago', data=df, palette='Reds')
plt.title('Ventas por Método de Pago')
plt.xlabel('Método de Pago')
plt.ylabel('Número de Ventas')
plt.xticks(rotation=45)
plt.show()

# --- Ejemplo de Gráfico 3: Ingresos Totales por Fecha (Serie de Tiempo) ---
# Agrupar por fecha y sumar los valores
ventas_por_fecha = df.groupby('fecha')['valor'].sum().reset_index()

plt.figure(figsize=(14, 7))
sns.lineplot(x='fecha', y='valor', data=ventas_por_fecha, marker='o')
plt.title('Ingresos Totales por Día')
plt.xlabel('Fecha')
plt.ylabel('Ingresos Totales')
plt.grid(True)
plt.show()

# --- Ejemplo de Gráfico 4: Distribución de Valores de Venta (Histograma) ---
plt.figure(figsize=(10, 6))
sns.histplot(df['valor'], bins=20, kde=True, palette='Blues')
plt.title('Distribución de los Valores de Venta')
plt.xlabel('Valor de Venta')
plt.ylabel('Frecuencia')
plt.show()
```
